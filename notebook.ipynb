```python

# Import the beautifulsoup 
# and request libraries of python.
import requests
import bs4
import numpy as np
from time import sleep
from random import randint

import pandas as pd
import numpy as np
```

# scrapping

```python
# Make two strings with default google search URL
# 'https://google.com/search?q=' and
# our customized search keyword.
# Concatenate them
text = "pengguna+linux"
pages = np.arange(0, 40, 10)

list_soup = []
for i, page in enumerate(pages):
    url = f'https://www.google.com/search?q={text}&client=firefox-b-d&sxsrf=APq-WBtVUnKnJFJJJ8XtMFHz7Em1EQOciA:1650188431391&ei=j-BbYoW0F6ah4t4PrbmRqAM&start={page}&sa=N&ved=2ahUKEwjFt4_65pr3AhWmkNgFHa1cBDU4ChDy0wN6BAgBEDs&biw=1920&bih=963&dpr=1'

    # Fetch the URL data using requests.get(url),
    # store it in a variable, request_result.
    request_result=requests.get( url )
    
    print(i, page, request_result)

    # Creating soup from the fetched request
    soup = bs4.BeautifulSoup(request_result.text,
                             "html.parser")
    # print(soup
    
    list_soup.append(soup)
    
    sleep(randint(2,10))
```

```python
title_list = []
url_list = []
page_list = []

for i, soup in enumerate(list_soup):

    links = soup.find_all("a")

    # get url and website title
    for link in links:
        link_href = link.get('href')
        
        if "url?q=" in link_href and not "webcache" in link_href:
            title = link.find_all('h3')
            
            if len(title) > 0:
                url_web = link.get('href').split("?q=")[1].split("&sa=U")[0]
                title_web = title[0].getText()
                
                # print(url_web)
                # print(title_web)
                # print("------")
                
                title_list.extend([title_web])
                url_list.extend([url_web])
```

```python
df = pd.DataFrame({'rank':np.arange(0, len(title_list), 1),
                   'url':url_list,
                   "title":title_list})
df
```

```python
df.to_csv("dataset.csv", sep=';', index=False, quoting=1)
```
